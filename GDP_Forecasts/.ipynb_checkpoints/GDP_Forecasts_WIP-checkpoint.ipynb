```python
import numpy as np

def check_float(x):
    try:
        a = pd.to_numeric(x)
        if np.isnan(a):
            b = False
        else:
            b = isinstance(a,float)
    except:
        b = False
    return b

def check_star(x):
    try:
        if '*' in x:
            b = 1
        else:
            b=0
    except:
        b=0
    return b
```

```python
for k in df:
    for n in [0,1]:
        month = k.split('_')[0]
        year = k.split('_')[1]

        if month.capitalize() in calendar.month_name[2:13]:
            forecast_year = 'Forecast_{}'.format(pd.to_numeric(year)+n)
            df[k][n]['yr_forecast'] = pd.to_numeric(year)+n
            
        else:
            forecast_year='Forecast_{}'.format(pd.to_numeric(year)+n-1)
            df[k][n]['yr_forecast'] = pd.to_numeric(year)+n-1
            
        cols_ = list(df[k][n].columns)
        cols_ = [cols_[-1]] + cols_[:-1]
        df[k][n] = df[k][n][cols_]
       
        
        for i in df[k][n].columns[1:]:
            num = df[k][n][i].map(lambda x : check_float(x)).sum()
            if num > 0:
                df[k][n].rename(columns = {i:forecast_year}, inplace = True)
                break
            else:
                    pass

        cols = [col for col in  df[k][n] if col.startswith('Forecast_')]
        df[k][n]=df[k][n].loc[:,:cols[0]].copy()
       
```

```python
for k in df:
    for n in [0,1]:   
        for i in df[k][n].columns:
            stars = df[k][n][i].map(lambda x : check_star(x)).sum()
            if (stars > 2) & (list(df[k][n].columns).index(i) !=1):
                new_forecasts = i
            elif (stars > 2) & (list(df[k][n].columns).index(i)==1):
                 df[k][n]['new_forecasts']=''
            else:
                pass
        
        df[k][n].rename(columns = {new_forecasts:'new_forecasts'}, inplace = True)    

        if 'Unnamed: 1' in df[k][n].columns:
            df[k][n].rename(columns = {'Unnamed: 1':'month_of_forecast'}, inplace = True)
        elif 'Unnamed: 0' in df[k][n].columns:
            df[k][n].rename(columns = {'Unnamed: 0':'month_of_forecast'}, inplace = True)
        else:
            df[k][n]['month_of_forecast']=''
        df[k][n]['forecast_publication']=k 
        df[k][n] = df[k][n].loc[:,~df[k][n].columns.duplicated()].copy()
        df[k][n].rename(columns = {df[k][n].columns[1]:'Forecaster'}, inplace = True)
        try:
            df[k][n] = df[k][n].drop(columns = ['Unnamed: 0']).copy()
        except:
            pass
        try:
            df[k][n] = df[k][n].drop(columns = ['Unnamed: 2']).copy()
        except:
            pass
        df[k][n] = df[k][n].dropna(subset=['Forecaster']).reset_index(drop=True)
        df[k][n] = df[k][n].dropna(subset=df[k][n].columns[1:],how='all').reset_index(drop=True)
```

```python
month_abbrs = [' '+x for x in list(calendar.month_abbr)[1:]]
for k in df:
    for n in [0,1]: 
        for x in month_abbrs:
            for i in range(len(df[k][n])):
                if x in df[k][n]['Forecaster'][i]:
                    df[k][n]['Forecaster'][i]=df[k][n]['Forecaster'][i].split(x)[0]
                    df[k][n]['month_of_forecast'][i]=x
                else:
                    pass

for k in df:
    for n in [0,1]: 
        for x in month_abbrs:
            for i in range(len(df[k][n])):
                if isinstance(df[k][n]['new_forecasts'][i], float):
                    pass
                else:
                    if x in df[k][n]['new_forecasts'][i]:
                        df[k][n]['new_forecasts'][i]=df[k][n]['new_forecasts'][i].split(x)[0]
                        df[k][n]['month_of_forecast'][i]=x
                    else:
                        pass
```

```python
remove_list =['city forecasters','non-city','output gap','range of', 'scenario','European Commission (mild']
df_all['removals']=0
for i in range(len(df_all['Forecaster'])):
    count = 0
    for x in remove_list:
        if x in df_all['Forecaster'][i].lower():
            count =+1
    df_all['removals'][i]=count

df_all = df_all[df_all.removals==0]
```

```python

df_all.reset_index(drop=True, inplace=True)
df_all['yr_of_forecast'] = 0
df_all['mnth_of_forecast'] = 0

for i in range(len(df_all.month_of_forecast)):
    df_all.month_of_forecast[i] = str(df_all.month_of_forecast[i]).strip().replace('-',' ')
    for x in month_abbrs:
        if x.strip() in df_all.month_of_forecast[i]:
            df_all.mnth_of_forecast[i]=x

    if len(str(df_all.month_of_forecast[i]).strip().split(' '))>1:
        df_all.yr_of_forecast[i] =int(df_all.month_of_forecast[i].strip().split(' ')[1].replace("'",""))+2000
    else:
        if df_all.mnth_of_forecast[i]!=0:
            if list(calendar.month_abbr).index(df_all.mnth_of_forecast[i].strip()) <=list(calendar.month_name).index(df_all.forecast_publication[i].split('_')[0].capitalize()):
                df_all.yr_of_forecast[i]=df_all.forecast_publication[i].split('_')[1]
            else:
                df_all.yr_of_forecast[i]=int(df_all.forecast_publication[i].split('_')[1])-1

        elif isinstance(df_all.mnth_of_forecast[i],int) or isinstance(df_all.mnth_of_forecast[i],float):

            df_all.mnth_of_forecast[i]=list(calendar.month_abbr)[list(calendar.month_name).index(df_all.forecast_publication[i].split('_')[0].capitalize())]
            df_all.yr_of_forecast[i] = df_all.forecast_publication[i].split('_')[1]
        else:
            pass
```

```python
links_dict = make_links_dict(get_links(),get_sub_links(get_links()))
```

```python
import PyPDF4
from urllib.request import urlopen
from io import BytesIO

contents_dict = {}
for u in links_dict.keys():
    url = links_dict[u]
    remoteFile = urlopen(Request(url)).read()
    memoryFile = BytesIO(remoteFile)
    reader = PyPDF4.PdfFileReader(memoryFile)
    page = reader.getPage(2)
    text = page.extractText()
    contents = {}
    for i,x in enumerate(text.split('\n')):
        if x.find('GDP')>0:
            contents[x] = text.split('\n')[i+1]
        else:
            pass
    contents_dict[u]=contents
```
